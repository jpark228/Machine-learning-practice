최소제곱

```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97]) 
```

```python
mx = np.mean(x)
my = np.mean(y)
```

```python
diviosr = sum([(i - mx)**2 for i in x])
```

```python
divisor = sum([(i - mx)**2 for i in x])

def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d
dividend = top(x, mx, y, my)
```

```python
a = dividend / divisor
b = my - (mx*a)
```

```python
print("x의 평균값: ", mx)
print("y의 평균값: ", my)

print("분모: ", divisor)
print("분자: ", dividend)

print("기울기 a = ", a)
print("y 절편 b = ", b)
```

평균 제곱 오차

```python
import numpy as np

fake_a = 3
fake_b = 76

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
def predict(x):
  return fake_a * x + fake_b
  
predict_result = []
```

```python
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))
```

```python
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y - y_pred)**2)


print("평균 제곱 오차: " + str(mse(y, predict_result)))
```
