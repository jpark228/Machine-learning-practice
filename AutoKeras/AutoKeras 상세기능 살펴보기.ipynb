```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

from tensorflow.keras.utils import to_categorical
```

```python
import autokeras as ak
```

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(60000, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

y_train= to_categorical(y_train)
y_test = to_categorical(y_test)

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
ak_model = ak.ImageClassifier(overwrite=True, max_trials=1)
```

```python
ak_model.fit(x_train, y_train,epochs=5)
```

```python
ak_model.evaluate(x_test, y_test)
```

```python
input_node = ak.ImageInput()
output_node = ak.ImageBlock(
    block_type = "resnet",
    normalize=True,
    augment=False
)(input_node)
```

```python
output_node = ak.ClassificationHead()(output_node)
```

```python
ak_model2 = ak.AutoModel(
    inputs = input_node, outputs = output_node, overwrite=True, max_trials=1
)
ak_model2.fit(x_train, y_train, epochs=5)
```

```python
ak_model2.evaluate(x_test, y_test)
```
