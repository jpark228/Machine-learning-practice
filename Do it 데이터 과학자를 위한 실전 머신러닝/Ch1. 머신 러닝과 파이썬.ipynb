```python
import numpy as np
from sklearn.model_selection import train_test_split

X, y = np.arange(120).reshape((30, 4)), list(range(30))
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"X_test_scaled의 첫 5개 샘플: \n{np.array2string(X_test_scaled[:5, :], precision=3, floatmode='fixed')}")
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
print(
    f"X_train_scaled의 첫 5개 샘플:\n{np.array2string(X_train_scaled[:5, :], precision=3, floatmode='fixed')}"
)
```
